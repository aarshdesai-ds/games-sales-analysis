
## 3. Data Cleaning and Preprocessing

We perform the following cleaning steps:
- Drop rows with missing values in essential columns (`Name`, `Genre`, `Year_of_Release`, `Publisher`)  
- Convert `Year_of_Release` to integer  
- Convert and fill missing `User_Score` values  
- Replace missing values in `Developer` and `Rating` with `'Unknown'`  
- Drop columns with excessive missing data (`Critic_Score`, `Critic_Count`, `User_Count`)

```python
df.isnull().sum()
```

```python
# Drop rows with missing critical ID fields
df = df.dropna(subset=['Name', 'Genre', 'Year_of_Release', 'Publisher'])

# Convert types
df['Year_of_Release'] = df['Year_of_Release'].astype(int)

# Convert User Score and fill missing
df['User_Score'] = pd.to_numeric(df['User_Score'], errors='coerce')
df['User_Score'].fillna(df['User_Score'].mean(), inplace=True)

# Fill missing ratings/dev info with placeholder
df['Rating'].fillna('Unknown', inplace=True)
df['Developer'].fillna('Unknown', inplace=True)

# Optionally drop critic/user count if not used
```

```python
df.isnull().sum()
```

```python
# Drop columns with heavy missing values
df = df.drop(columns=['Critic_Score', 'Critic_Count', 'User_Count'])
```

## 22. 🔗 Pearson Correlation: User Score vs Global Sales

We compute the **Pearson correlation coefficient** between `User_Score` and `Global_Sales` to determine whether user feedback aligns with commercial performance.

### 🧠 Interpretation:
- A **positive value near 1**: Higher scores tend to mean higher sales.
- A **value near 0**: No clear linear relationship.
- A **negative value**: Higher scores relate to lower sales (unusual).
- A **low p-value (< 0.05)** means the relationship is statistically significant.

> This test helps validate whether game popularity among players translates into financial success.


```python
from scipy.stats import pearsonr

corr_coeff, p_value = pearsonr(df['User_Score'], df['Global_Sales'])
print(f"Pearson Correlation between User Score and Global Sales: {corr_coeff:.2f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("✅ Statistically significant: User Score and Global Sales are correlated.")
else:
    print("❌ Not statistically significant: No strong evidence of correlation.")
```

## 23. ⚔️ T-Test: Action vs Strategy Game Sales

We run a **two-sample t-test** to compare the average `Global_Sales` of Action vs Strategy games.

### 🔬 Hypotheses:
- **H₀ (null)**: Both genres have the same average sales.
- **H₁ (alternative)**: Their average sales are significantly different.

### ✅ Why this matters:
- Confirms whether some genres consistently outperform others in revenue.
- Can guide publishers on genre investment decisions.

> We use Welch’s t-test (`equal_var=False`) since variance in sales likely differs between these genres.


```python
from scipy.stats import ttest_ind

action_sales = df[df['Genre'] == 'Action']['Global_Sales']
strategy_sales = df[df['Genre'] == 'Strategy']['Global_Sales']

t_stat, p_val = ttest_ind(action_sales, strategy_sales, equal_var=False)
print("\nT-Test: Action vs Strategy Game Sales")
print(f"T-statistic: {t_stat:.2f}, P-value: {p_val:.4f}")

if p_val < 0.05:
    print("✅ Significant: Action games and Strategy games have different average sales.")
else:
    print("❌ Not significant: No strong difference in sales between Action and Strategy games.")
```

## 24. 🧠 ANOVA: Do Ratings Affect Global Sales?

Using a one-way **ANOVA (Analysis of Variance)**, we check if ESRB `Rating` categories (like `E`, `T`, `M`, `AO`) are associated with differences in `Global_Sales`.

### 🔍 Hypotheses:
- **H₀**: All groups have the same mean sales.
- **H₁**: At least one group’s mean is different.

### 🎯 Business Use:
- Confirms if content rating influences a game’s commercial performance.
- May reveal that certain ratings (e.g., `M` for Mature) limit or boost market reach.

> A significant result (p < 0.05) implies that ESRB ratings impact how much a game sells.


```python
from scipy.stats import f_oneway

rating_groups = [group['Global_Sales'].values for name, group in df.groupby('Rating')]
f_stat, p_val = f_oneway(*rating_groups)

print("\nANOVA Test across Ratings:")
print(f"F-statistic: {f_stat:.2f}, P-value: {p_val:.4f}")

if p_val < 0.05:
    print("✅ Significant: At least one rating group differs in average Global Sales.")
else:
    print("❌ Not significant: Sales don't vary significantly across rating categories.")
```

## 25. 🎲 Chi-Squared Test: Genre vs Rating Dependency

We conduct a **Chi-squared test of independence** to determine whether there is a **relationship between `Genre` and `Rating`**.

### 🔍 Hypotheses:
- **H₀**: `Genre` and `Rating` are independent (e.g., any genre can have any rating).
- **H₁**: They are dependent (e.g., Shooters are more likely to be `M`, Family games are often `E`).

### 📘 Usefulness:
- Shows how content restrictions or audience targeting correlate with genre types.
- Informs how regulatory and cultural norms shape the market for specific game genres.

> A significant p-value (typically < 0.05) supports the idea that game type influences or is influenced by its age-appropriateness classification.


```python
# 26. 🎲 Chi-Squared Test: Are Genre and Rating Independent?

import pandas as pd
from scipy.stats import chi2_contingency

# Create a contingency table
contingency_table = pd.crosstab(df['Genre'], df['Rating'])

# Perform Chi-squared test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print("Chi-squared Test between Genre and Rating")
print(f"Chi2 Statistic: {chi2:.2f}")
print(f"Degrees of Freedom: {dof}")
print(f"P-value: {p:.4f}")

if p < 0.05:
    print("✅ Statistically significant: Genre and Rating are NOT independent.")
else:
    print("❌ Not statistically significant: No evidence of dependency between Genre and Rating.")
```

- **Sports**, **Shooter**, and **Role-Playing** games also contribute significantly to overall sales.
- Niche genres like **Strategy**, **Puzzle**, and **Adventure** show lower sales but maintain consistent presence.

### 🕹️ Platforms
- **PS2**, **X360**, and **Wii** are among the top-selling platforms, with PS2 showing exceptional dominance.
- Platform-specific trends suggest the importance of console cycles in determining sales peaks.

### 🌍 Regional Markets
- **North America** accounts for nearly 50% of total sales, making it the most lucrative region.
- **Europe** and **Japan** follow, with each showing unique genre/platform preferences.
- Other regions contribute a smaller but growing share of the market.

### ⏳ Temporal Trends
- Game releases and sales **peaked around 2008–2010**, followed by a noticeable decline.
- This decline could be due to the rise of mobile gaming, digital-only titles, or data limitations.

### 🌟 User Scores
- Most user ratings cluster between **6 and 8**, suggesting favorable but conservative feedback from players.
- The user score distribution supports the notion of a competitive market with few outstanding outliers.

### 🧪 Statistical Insights
- **Sales distributions are right-skewed**, with only a small fraction of games achieving blockbuster status.
- **Chi-squared and ANOVA tests** indicate that genre and rating have statistically significant influence on sales.
- **User Score** does not strongly correlate with sales, suggesting other factors (e.g., marketing, brand) are more influential.
